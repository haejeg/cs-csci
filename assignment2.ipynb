```python
import numpy as np
X = np . array ([
[0.5 , 3] , [2 , 2.2] , [3 , 3.5] ,
[10 , 9] , [11.5 , 10.5] , [9.5 , 11.2]
])
```

```python
import sklearn
```

```python
# Part A (Initialiing K Means)
kmeans = sklearn.cluster.KMeans(n_clusters=2, random_state=42, n_init=10).fit(X)
print("Cluster Labels")
print(kmeans.labels_)
print("\nCluster Centers")
print(kmeans.cluster_centers_)
```

```python
# Part B (Inertia, WCSS)
print(kmeans.inertia_)
```

WCSS is the sum of squared distances of all points and their assigned cluster centroids.

```python
# Part C (Graphing)
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, label="points")
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], marker='*', label="centroids")
plt.legend()
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Plot of Clusters and Centroids for Given Data')
plt.show()
```

```python
# Part D (Elbow Style Check)
kMeansList = []
for k in range(1, 5):
  newKMeans = sklearn.cluster.KMeans(n_clusters=k, random_state=42, n_init=10).fit(X)
  kMeansList.append(newKMeans)

plt.plot(range(1, 5), [kmeans.inertia_ for kmeans in kMeansList])
plt.xlabel('K Number of Clusters')
plt.ylabel('WCSS')
plt.title('Elbow Style Check')
```

For the given set of data points, I believe that k = 2 is a reasonable choice for this dataset, since the elbow-style indicates that adding more clusters beyond k = 2 returns diminishing returns in the reduction of WCSS.
